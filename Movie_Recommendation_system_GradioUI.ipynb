```python
df['tags']=df['genre']+df['overview']
```

New Dataset

```python
df1=df[['id','title','tags']]
```

Now converting the tags into vectors.


```python
df1.tail(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(df1['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

Similarity Calculation.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos=cosine_similarity(vector)
```

```python
cos
```

```python
df1[df1['title']=='The Godfather']
```

Now defining as function for recommending movies.


```python
def recommendator(movie):
  index=df1[df1['title']==movie].index[0]
  d=sorted(list(enumerate(cos[index])), reverse=True, key=lambda vector: vector[1])
  for i in d[0:6]:
    print(df1.iloc[i[0]].title)
```

```python
def recommendatorgr(movie):
    recommendations = []
    index = df1[df1['title'] == movie].index
    if index.empty:
        recommendations.append("Movie not found in the database.")
    else:
        index = index[0]
        d = sorted(list(enumerate(cos[index])), reverse=True, key=lambda vector: vector[1])
        for i in d[1:6]:
            recommendations.append(df1.iloc[i[0]].title)
    return "\n".join(recommendations)
```

```python
recommendatorgr('Captain America')
```

```python
import gradio as gr

IF = gr.Interface(
    fn=recommendatorgr,
    inputs=gr.components.Textbox(lines=2, placeholder="Enter Movie Name"),
    outputs="text",
    title="Movie Recommendation System"
)
IF.launch()
```
